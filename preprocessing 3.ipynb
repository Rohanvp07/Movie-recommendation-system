## Extracting features of 2021(upto June) movies from Wikipedia

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2021"
df1 = pd.read_html(link, header=0)[3]
df2 = pd.read_html(link, header=0)[4]

df_2021 = df1.append(df2,ignore_index=True)
df_2021
```

```python
from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = '49a9ec3add7d7c5d33c1bfb4485ab694'   #api key
```

```python
from tmdbv3api import Movie
tmdb_movie = Movie()
def get_genre(x):
    try:
        genres = []
        result = tmdb_movie.search(x)
        movie_id = result[0].id
        response = requests.get('https://api.themoviedb.org/3/movie/{}?api_key={}'.format(movie_id,tmdb.api_key))
        data_json = response.json()
    finally:
        if data_json['genres']:
            genre_str = " " 
            for i in range(0,len(data_json['genres'])):
                genres.append(data_json['genres'][i]['name'])
            return genre_str.join(genres)
        else:
            return np.nan
```

```python
#getting the genres of the movies
from tqdm import tqdm
for i,data in tqdm(df_2018.iterrows()):
    df_2018.at[i,'genres']=get_genre(data["Title"])
```

```python
for i,data in tqdm(df_2019.iterrows()):
    df_2019.at[i,'genres']=get_genre(data["Title"])
```

```python
df = df[['Title','Cast and crew','genres']]
df
```

```python
def get_director(x):           #getting the director names of the movies
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    elif " (director/screenplay)" in x:
        return x.split(" (director/screenplay)")[0]
    else:
        return np.nan
```

```python
for i,data in df.iterrows():
    df.at[i,"director_name"]=get_director(data["Cast and crew"])
```

```python
df.head()
```

```python
def get_actors(x):                                      #getting the 3 actor names of the movies
    casts=(x.split("screenplay); ")[-1]).split(", ")
    if len(casts)>=3:
        return casts[0],casts[1],casts[2]
    elif len(casts)==2:
        casts.extend([np.nan,np.nan])
    elif len(casts)==1:
        casts.extend([np.nan,np.nan,np.nan])
    return casts[0],casts[1],casts[2]
```

```python
for i,data in df.iterrows():
    a1,a2,a3=get_actors(data["Cast and crew"])
    df.at[i,"actor_1_name"]=a1
    df.at[i,"actor_2_name"]=a2
    df.at[i,"actor_3_name"]=a3
```

```python
df = df[['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
df
```

```python
df.isna().sum()         #searching for na values if any
```

```python
#replacing the na values with an empty string

df['director_name'] = df['director_name'].replace(np.nan, '')
df['actor_2_name'] = df['actor_2_name'].replace(np.nan, '')
df['actor_3_name'] = df['actor_3_name'].replace(np.nan, '')
df['genres'] = df['genres'].replace(np.nan, '')
```

```python
df['movie_title'] = df['movie_title'].str.lower()
```

```python
df['bow'] = df['actor_1_name'] + ' ' + df['actor_2_name'] + ' '+ df['actor_3_name'] + ' '+ df['director_name'] +' ' + df['genres']
```
