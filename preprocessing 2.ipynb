```python
import numpy as np
import pandas as pd
```

```python
credits = pd.read_csv('credits.csv')
```

```python
credits.head(10)
```

```python
meta = pd.read_csv('movies_metadata.csv')
```

```python
meta.head()
```

```python
meta.columns
```

```python
meta['release_date'] = pd.to_datetime(meta['release_date'], errors='coerce')
```

```python
meta['year'] = meta['release_date'].dt.year
```

```python
meta['year'].value_counts().sort_index()
```

```python
# Getting only 2017 movies as we already have movies up to the year 2016 in preprocessing 1 file. 
# We don't have enough data for the movies from 2018, 2019 and 2020. 
# We'll deal with it in the upcoming preprocessing files
new_meta = meta[['genres','id','title','year']][meta.year==2017]
```

```python
new_meta
```

```python
new_meta['id'] = new_meta['id'].astype(int)
```

```python
#we are merging the data to get the details like actors, director and the genres of the movies
#merging the data on the id column, ie, getting the details of 2017 movies only
data = pd.merge(new_meta, credits, on='id')      
```

```python
data.head()
```

```python
# evaluates an expression node or a string containing a Python literal or container display
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
data.genres
```

```python
def make_genresList(x):    #getting the genres of the movies
    gen = []
    for i in x:
        if i['name'] == 'Science Fiction':
            gen.append('Sci-Fi')
        else:
            gen.append(i['name'])
    if gen == []:
        return np.NaN
    else:
        return (" ".join(gen))
```

```python
for i,d in data.iterrows():
    data.at[i,'genres']=make_genresList(d["genres"])
```

```python
data['genres']
```

```python
def get_actors(x):     #getting the actors of the movies
    casts = []
    c=0
    for i in x:
        casts.append(i['name'])
        c+=1
        if c==3:
            break
    if len(casts)==3:
        return casts
    elif len(casts)==2:
        casts.append(np.NaN)
    elif len(casts)==1:
        casts.extend([np.NaN,np.NaN])
    elif len(casts)==0:
        casts.extend([np.NaN,np.NaN,np.NaN])
    return casts
```

```python
for i,d in data.iterrows():
    actors=get_actors(data["cast"][i])
    data.at[i,'actor_1_name']=actors[0]
    data.at[i,'actor_2_name']=actors[1]
    data.at[i,'actor_3_name']=actors[2]
```

```python
data['actor_1_name']
```

```python
data['actor_2_name']
```

```python
data['actor_3_name']
```

```python
def get_director(x):
    dt = []
    for i in x:
        if i['job'] == 'Director':
            dt.append(i['name'])
    if dt == []:
        return np.NaN
    else:
        return (" ".join(dt))
```

```python
for i,d in data.iterrows():
    data.at[i,'director_name']=get_director(d["crew"])
```

```python
data['director_name']
```

```python
movie = data[['director_name','actor_1_name','actor_2_name','actor_3_name','genres','title']]
```

```python
movie = movie.rename(columns={'title':'movie_title'})
```

```python
movie.head(10)
```

```python
movie.isna().sum()
```

```python
movie['actor_1_name'] = movie['actor_1_name'].replace(np.nan, '')
movie['actor_2_name'] = movie['actor_2_name'].replace(np.nan, '')
movie['actor_3_name'] = movie['actor_3_name'].replace(np.nan, '')
movie['director_name'] = movie['director_name'].replace(np.nan, '')
movie['genres'] = movie['genres'].replace(np.nan, '')
movie['movie_title'] = movie['movie_title'].replace(np.nan, '')
```

```python
movie.isna().sum()
```

```python
movie['movie_title'] = movie['movie_title'].str.lower()
```

```python
movie['bow'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' '+ movie['actor_3_name'] + ' '+ movie['director_name'] +' ' + movie['genres']
```
