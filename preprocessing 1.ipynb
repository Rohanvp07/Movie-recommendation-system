```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('movie_metadata.csv')
```

```python
# we have movies only upto 2016
import matplotlib.pyplot as plt
data.title_year.value_counts(dropna=False).sort_index().plot(kind='barh',figsize=(15,16))
plt.title("How many movies were released per year!")
plt.show()
```

```python
data.columns
```

```python
# recommendation will be based on these features only
data = data[['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
data.head(10)
```

```python
#replacing the null values for these columns with empty string
data['actor_1_name'] = data['actor_1_name'].replace(np.nan, '')
data['actor_2_name'] = data['actor_2_name'].replace(np.nan, '')
data['actor_3_name'] = data['actor_3_name'].replace(np.nan, '')
data['director_name'] = data['director_name'].replace(np.nan, '')
```

```python
data.head(10)
```

```python
data['genres'] = data['genres'].str.replace('|', ' ')
```

```python
data.head(10)
```

```python
data['movie_title'] = data['movie_title'].str.lower()
```

```python
# null terminating char at the end
data['movie_title'][0]
```

```python
# removing the null terminating char at the end
for i,d in data.iterrows():
    data.at[i,"movie_title"]=d["movie_title"][:-1]
```
